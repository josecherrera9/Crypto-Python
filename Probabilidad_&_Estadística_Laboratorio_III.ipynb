```python
import pandas as pd
import matplotlib.pyplot as plt
```

Importando DataBase

**Pruebas de hipótesis**

***Temperatura***
*   Hipotesis nula: La temperatura media es de 8 grados
*   Hipótesis alternativa: La temperatura es mayor o menor

```python
print("La media de la temperatura es: ", round(df["Temperature"].mean(), 3))
print("La temperatura tiene una desviación estandar de: ", round(df["Temperature"].std(), 3))
```

Dado que se busca una hipótesis con respecto a la media y que es conocida esta y su desviación
estandar se utilziará una prueba Z

```python
prueba_z = (len(df["Temperature"]) - df["Temperature"].mean()) / df["Temperature"].std()
round(prueba_z, 3)
```

***Ventas***


*   Hipótesis nula: La mayoría de los días las ventas son mayores a 60
*   Hipótesis alternativa: Se obtuvieron ventas menores a 60 en la mayoría de los días



```python
print("La media de las ventas es: ", df["Sales"].mean())
print("Las ventas tienen una desviación estandar de: ", round(df["Sales"].std(), 3))
```

Se utiliza una prueba Z

```python
prueba_z2 = (len(df["Sales"])) - df["Sales"].mean() / df["Sales"].std()
round(prueba_z2, 3)
```

***Ingresos***


*   Hipótesis nula: Se obtuvieron ingresos mayores a 220 en la mayoría de los días
*   Hipótesis alternativa: Se obtuvieron ingresos menores a 220 en la mayoría de los días



```python
print("La media de ingresos es de: ", df["GrossIncome"].mean())
print("Los ingresos tienen una desviación estandar de: ", round(df["GrossIncome"].std(), 3))
```
