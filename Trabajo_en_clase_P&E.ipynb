```python
import numpy as np
import matplotlib.pyplot as plt
```

**Pregunta 1:**

```python
uniforme = np.random.uniform(size=10000)
```

**Pregunta 2:**

```python
normal = np.random.normal(size=10000)
print("La media es de: ", round(normal.mean(), 4))
print("La desviación estandar es de: ", round(normal.std(), 4))
```

**Pregunta 3:**

```python
eje_y1 = normal[0:25]
plt.hist(eje_y1, color="orange", ec="red")
plt.title("Tamaño de muestra: 25")
plt.xlabel("Rango de números")
plt.ylabel("Cantidad de números")
plt.show()

eje_y2 = normal[0:50]
plt.hist(eje_y2, color="skyblue", ec="blue")
plt.title("Tamaño de muestra: 50")
plt.xlabel("Rango de números")
plt.ylabel("Cantidad de números")
plt.show()

eje_y3 = normal[0:100]
plt.hist(eje_y3, color="purple", ec="orchid")
plt.title("Tamaño de muestra: 100")
plt.xlabel("Rango de números")
plt.ylabel("Cantidad de números")
plt.show()

eje_y4 = normal[0:900]
plt.hist(eje_y4, ec="skyblue")
plt.title("Tamaño de muestra: 900")
plt.xlabel("Rango de números")
plt.ylabel("Cantidad de números")
plt.show()
```

**Respuesta:** Se puede ver que conforme aumenta la muestra se forma una gráfica de distribución normal

**Pregunta 4:**

```python
lista_de_medias = []
lista_de_varianzas = []
for i in range(100):
  medias = np.random.choice(uniforme, size=10, replace=True, p=None).mean()
  varianzas = np.random.choice(uniforme, size=10, replace=True, p=None).var()
  lista_de_medias.append(medias)
  lista_de_varianzas.append(varianzas)
```

**Pregunta 5:**

```python
plt.hist(lista_de_medias, color="salmon", ec="red")
plt.xlabel("Rangos")
plt.ylabel("Valor de las medias")
plt.title("Medias de las muestras")
plt.show()

plt.hist(lista_de_varianzas, color="seagreen", ec="palegreen")
plt.xlabel("Rangos")
plt.ylabel("Valor de las varianzas")
plt.title("Varianzas de las muestras")
plt.show()
```

**Pregunta 6:**

```python
print("Media de las medias: ", round(np.mean(lista_de_medias), 4))
print("Varianza de las medias: ", round(np.var(lista_de_medias), 4))
```

**Pregunta 7:**

```python
print("Media de las varianzas: ", round(np.mean(lista_de_varianzas), 4))
print("Varianza de las varianzas: ", round(np.var(lista_de_varianzas), 4))
```

**Pregunta 8:**

```python
lista_de_medias = []
lista_de_varianzas = []
for i in range(100):
  medias = np.random.choice(normal, size=10, replace=True, p=None).mean()
  varianzas = np.random.choice(normal, size=10, replace=True, p=None).var()
  lista_de_medias.append(medias)
  lista_de_varianzas.append(varianzas)

plt.hist(lista_de_medias, color="salmon", ec="red")
plt.xlabel("Rangos")
plt.ylabel("Valor de las medias")
plt.title("Medias de las muestras")
plt.show()

plt.hist(lista_de_varianzas, color="seagreen", ec="palegreen")
plt.xlabel("Rangos")
plt.ylabel("Valor de las varianzas")
plt.title("Varianzas de las muestras")
plt.show()

print()
print("Media de las medias: ",round(np.mean(lista_de_medias), 4))
print("Varianza de las medias: ", round(np.var(lista_de_medias), 4))
print()
print("Media de las varianzas: ", round(np.mean(lista_de_varianzas), 4))
print("Varianza de las varianzas: ", round(np.var(lista_de_varianzas), 4))
```

**Respuesta:** Se ve que el histograma de las medias sigue una distribución normal y el de las medias una exponencial.
