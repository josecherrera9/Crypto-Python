# **Imports**

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas_datareader import data as pdr
```

# **Extracción**

```python
#Extracción de datos

def extraer():
  website = "https://www.coingecko.com/es"

  response = requests.get(website)

  soup = BeautifulSoup(response.content, "html.parser")

  results = soup.find("table", {"class":"table-scrollable"}).find("tbody").find_all("tr")


#Loop para aplicar a todos los resultados

  name = []
  price = []
  change_1h = []
  change_24h = []
  change_7d = []
  volume_24h = []
  market_cap = []

  for result in results:

#Variables globales
    global nombre
  
    global precio

    global cambio_1h

    global cambio_24h

    global cambio_7d

    global volumen_24h

    global coin_market_cap

    global crypto_data_full_table

#Nombre
    nombre = name.append(result.find("a",{"class": "tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between"}).get_text().strip())

#Precio
    precio = price.append(result.find("td",{"class":"td-price"}).get_text().strip())

#Change 1h
    cambio_1h = change_1h.append(result.find("td",{"class":"td-change1h"}).get_text().strip())

#Change 24h
    cambio_24h = change_24h.append(result.find("td",{"class":"td-change24h"}).get_text().strip())

#Change 7d
    cambio_7d = change_7d.append(result.find("td",{"class":"td-change7d"}).get_text().strip())

#Volumen 24h
    volumen_24h = volume_24h.append(result.find("td",{"class":"td-liquidity_score"}).get_text().strip())

#Market Cap
    coin_market_cap = market_cap.append(result.find("td",{"class":"td-market_cap"}).get_text().strip())


#Generación de DataFrame
  crypto_data_nombre = pd.DataFrame({"Moneda": name})

  crypto_data_precio = pd.DataFrame({"Moneda": name, "Precio" : price})

  crypto_data_cambio_1h = pd.DataFrame({"Moneda" : name, "Cambio 1h" : change_1h})

  crypto_data_cambio_24h = pd.DataFrame({"Moneda" : name, "Cambio 24h" : change_24h})

  crypto_data_cambio_7d = pd.DataFrame({"Moneda" : name, "Cambio 7d" : change_7d})

  crypto_data_volumen_24h = pd.DataFrame({"Moneda" : name, "Volumen 24 h" : volume_24h})

  crypto_data_market_cap = pd.DataFrame({"Moneda" : name, "Market Cap" : market_cap})

  crypto_data_full_table = pd.DataFrame({"Moneda": name, "Precio" : price, "Cambio 1h" : change_1h, "Cambio 24h" : change_24h, 
                                         "Cambio 7d" : change_7d, "Volumen 24 h" : volume_24h, "Market Cap" : market_cap})

  return crypto_data_precio
```
